```python
def buscar(self, **campos):
    '''
    Busca entre los registros de la Agenda que coincidan con los campos dados.
    Devuelve una lista de diccionarios, donde cada diccionario corresponde a un
    registro que cumple con los criterios de búsqueda
    
    :param campos: diccionarios con el nombre del campo como clave y el dato a buscar como valor.
    '''
    resultados = []
    with open(self._archivo, "rb") as fd:
        for i in range(self._cant_registros):
            fd.seek(i * self._len_registro)
            registro = fd.read(self._len_registro)
            valores = struct.unpack(self._formato, registro)
            registro_dict = {campo: valor.decode().strip() for campo, valor in zip(self._campos.keys(), valores)}
                
            # Se verifica si el registro cumple con todos los criterios de búsqueda
            if all(registro_dict.get(campo) == valor for campo, valor in campos.items()):
                resultados.append(registro_dict)
                    
    return resultados
```

````python
import os
import struct


class AgendaException(Exception):
    """Clase base para excepciones de Agenda"""

    pass


class Agenda:
    """Agenda con registros de tamaño fijo
    y campos de longitud fija
    """

    def __init__(self, archivo, **campos):
        """Constructor de la clase Agenda
        :param archivo: nombre del archivo donde se guardan los registros
        :param campos: diccionario con los campos de la agenda y su longitud
        """
        self._archivo = archivo
        self._campos = campos  # diccionario con los campos de la agenda y su longitud

        # "30s30s20s" -> 30 caracteres para el primer campo, 30 para el segundo y 20 para el tercero
        self._formato = "".join(f"{longitud}s" for longitud in campos.values())
        self._len_registro = struct.calcsize(self._formato)

        try:
            # obtengo el tamaño del archivo si es mayor que cero ya teniamos registros guardados
            tam_archivo = os.path.getsize(archivo)
            self._cant_registros = tam_archivo // self._len_registro
        except FileNotFoundError:
            self._cant_registros = 0

    def guardar_contacto(self, **datos):
        campos_validos = set(self._campos.keys())
        campos_recibidos = set(datos.keys())

        if campos_validos != campos_recibidos:
            raise AgendaException("Los campos recibidos no coinciden con los campos definidos en la agenda.")

        with open(self._archivo, "ab") as fd:
            valores = [campo.encode() for campo in datos.values()]
            fd.write(struct.pack(self._formato, *valores))
        self._cant_registros += 1

    def cantidad_registros(self):
        return self._cant_registros

    def campos(self):
        return self._campos

    def __iter__(self):
        return AgendaIterator(self)

    def buscar(self, **campos):

        resultados = []
        with open(self._archivo, "rb") as fd:
            for i in range(self._cant_registros):
                fd.seek(i * self._len_registro)
                registro = fd.read(self._len_registro)
                valores = struct.unpack(self._formato, registro)
                registro_dict = {campo: valor.decode().strip() for campo, valor in zip(self._campos.keys(), valores)}
                
                # Se verifica si el registro cumple con todos los criterios de búsqueda
                if all(registro_dict.get(campo) == valor for campo, valor in campos.items()):
                    resultados.append(registro_dict)
                    
        return resultados


class AgendaIterator:
    def __init__(self, agenda):
        self._agenda = agenda
        self._index = 0

    def __next__(self):
        if self._index >= self._agenda._cant_registros:
            raise StopIteration

        with open(self._agenda._archivo, "rb") as fd:
            posicion = self._index * self._agenda._len_registro
            fd.seek(posicion)
            registro = fd.read(self._agenda._len_registro)
            self._index += 1
            if len(registro) == self._agenda._len_registro:
                valores = struct.unpack(self._agenda._formato, registro)
                resultado = tuple(valor.decode() for valor in valores)
            else:
                raise AgendaException("Error al leer el registro.")

        return resultado
````

**Ej 2:** Modificar la Agenda con registro de tamaño fijo, cantidad de campos fija y longitud de campos variable vista en clase para usar un separador entre campos, como se hace con las archivos CSV, en lugar de prefijar cada campo con su longitud.

````python
import os

class AgendaException(Exception):
    """Clase base para excepciones de Agenda"""
    pass


class Agenda:
    """Agenda con registros de tamaño fijo y separador de campos"""

    def __init__(self, archivo, **campos):
        """
        Constructor de la clase Agenda.
        :param archivo: Nombre del archivo donde se guardan los registros.
        :param campos: Diccionario con los campos de la agenda.
        """
        self._archivo = archivo
        self._campos = list(campos.keys())  # lista con los nombres de los campos de la agenda
        self._separador = ','

        # Revisa si el archivo ya contiene datos, y si es así, cuenta los registros existentes.
        try:
            with open(archivo, "r", encoding="utf-8") as f:
                self._cant_registros = sum(1 for _ in f)
        except FileNotFoundError:
            self._cant_registros = 0

    def guardar_contacto(self, **datos):
        campos_validos = set(self._campos)
        campos_recibidos = set(datos.keys())

        if campos_validos != campos_recibidos:
            raise AgendaException("Los campos recibidos no coinciden con los campos definidos en la agenda.")

        with open(self._archivo, "a", encoding="utf-8") as fd:
            valores = [str(datos[campo]).replace(self._separador, " ") for campo in self._campos]
            fd.write(self._separador.join(valores) + "\n")
        self._cant_registros += 1

    def cantidad_registros(self):
        return self._cant_registros

    def campos(self):
        return self._campos

    def __iter__(self):
        return AgendaIterator(self)

    def buscar(self, **campos):
        resultados = []
        with open(self._archivo, "r", encoding="utf-8") as fd:
            for linea in fd:
                valores = linea.strip().split(self._separador)
                registro_dict = {campo: valor for campo, valor in zip(self._campos, valores)}

                # Verifica si el registro cumple con todos los criterios de búsqueda
                if all(registro_dict.get(campo) == str(valor) for campo, valor in campos.items()):
                    resultados.append(registro_dict)

        return resultados


class AgendaIterator:
    def __init__(self, agenda):
        self._agenda = agenda
        self._index = 0
        self._file = open(self._agenda._archivo, "r", encoding="utf-8")

    def __next__(self):
        linea = self._file.readline()
        if not linea:
            self._file.close()
            raise StopIteration

        valores = linea.strip().split(self._agenda._separador)
        return tuple(valores)
````

**Ej 3:**  Programar una nueva clase Agenda que soporte registros de longitud variable con cantidad de campos fija, es decir la longitud de los campos también debe ser variable. Cada registro y cada campo debe estar precedido por su longitud, para lo cual se deben utilizar 2 bytes.

````python
import os
import struct

class AgendaException(Exception):
    """Clase base para excepciones de Agenda"""
    pass


class Agenda:
    """Agenda con registros de longitud variable y cantidad fija de campos"""

    def __init__(self, archivo, **campos):
        """
        Constructor de la clase Agenda.
        :param archivo: Nombre del archivo donde se guardan los registros.
        :param campos: Diccionario con los nombres de los campos de la agenda.
        """
        self._archivo = archivo
        self._campos = list(campos.keys())  # lista con los nombres de los campos de la agenda

        # Verificar si el archivo ya contiene registros y contar el total
        try:
            with open(archivo, "rb") as f:
                self._cant_registros = self._contar_registros(f)
        except FileNotFoundError:
            self._cant_registros = 0

    def _contar_registros(self, file):
        count = 0
        while True:
            inicio = file.read(2)
            if not inicio:
                break
            longitud_registro, = struct.unpack("H", inicio)
            file.seek(longitud_registro, os.SEEK_CUR)
            count += 1
        return count

    def guardar_contacto(self, **datos):
        campos_validos = set(self._campos)
        campos_recibidos = set(datos.keys())

        if campos_validos != campos_recibidos:
            raise AgendaException("Los campos recibidos no coinciden con los campos definidos en la agenda.")

        with open(self._archivo, "ab") as fd:
            contenido_registro = b""
            for campo in self._campos:
                valor_bytes = datos[campo].encode("utf-8")
                longitud_campo = len(valor_bytes)
                contenido_registro += struct.pack("H", longitud_campo) + valor_bytes
            
            longitud_registro = len(contenido_registro)
            fd.write(struct.pack("H", longitud_registro) + contenido_registro)
        
        self._cant_registros += 1

    def cantidad_registros(self):
        return self._cant_registros

    def campos(self):
        return self._campos

    def __iter__(self):
        return AgendaIterator(self)

    def buscar(self, **campos):
        resultados = []
        with open(self._archivo, "rb") as fd:
            while True:
                inicio = fd.read(2)
                if not inicio:
                    break
                longitud_registro, = struct.unpack("H", inicio)
                contenido_registro = fd.read(longitud_registro)

                # Leer los campos con sus longitudes
                offset = 0
                registro_dict = {}
                for campo in self._campos:
                    longitud_campo, = struct.unpack("H", contenido_registro[offset:offset+2])
                    offset += 2
                    valor = contenido_registro[offset:offset+longitud_campo].decode("utf-8")
                    registro_dict[campo] = valor
                    offset += longitud_campo

                # Verificar si el registro coincide con los criterios de búsqueda
                if all(registro_dict.get(campo) == str(valor) for campo, valor in campos.items()):
                    resultados.append(registro_dict)

        return resultados


class AgendaIterator:
    def __init__(self, agenda):
        self._agenda = agenda
        self._file = open(self._agenda._archivo, "rb")

    def __next__(self):
        inicio = self._file.read(2)
        if not inicio:
            self._file.close()
            raise StopIteration

        longitud_registro, = struct.unpack("H", inicio)
        contenido_registro = self._file.read(longitud_registro)

        # Leer los campos con sus longitudes
        offset = 0
        registro = []
        for campo in self._agenda._campos:
            longitud_campo, = struct.unpack("H", contenido_registro[offset:offset+2])
            offset += 2
            valor = contenido_registro[offset:offset+longitud_campo].decode("utf-8")
            registro.append(valor)
            offset += longitud_campo

        return tuple(registro)
````

**Ej 4:** Crear un programa para gestionar la tabla de posiciones de un torneo de futbol. Los datos se deben persistir en un archivo llamado torneo.csv con las siguientes columnas: Fecha-Partido, Equipo Local, Goles Local, Equipo Visitante, Goles Visitante. El programa debe poder gestionar alta, búsqueda y modificaciones de partidos. Por ejemplo debe permitir buscar todos los partidos que se jugaron en determinada fecha, o los partidos en lo que jugo un equipo determinado, indicando la condición de local o visitante. 

````python
import csv
import os

class TorneoFutbol:
    def __init__(self, archivo="torneo.csv"):
        self.archivo = archivo
        self._verificar_archivo()

    def _verificar_archivo(self):
        # Crea el archivo si no existe y escribe los encabezados
        if not os.path.exists(self.archivo):
            with open(self.archivo, mode="w", newline="") as file:
                writer = csv.writer(file)
                writer.writerow(["Fecha-Partido", "Equipo Local", "Goles Local", "Equipo Visitante", "Goles Visitante"])

    def agregar_partido(self, fecha, equipo_local, goles_local, equipo_visitante, goles_visitante):
        """Agregar un nuevo partido al archivo."""
        with open(self.archivo, mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([fecha, equipo_local, goles_local, equipo_visitante, goles_visitante])

    def buscar_partidos_por_fecha(self, fecha):
        """Buscar partidos por fecha."""
        resultados = []
        with open(self.archivo, mode="r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                if row["Fecha-Partido"] == fecha:
                    resultados.append(row)
        return resultados

    def buscar_partidos_por_equipo(self, equipo, condicion="ambos"):
        """
        Buscar partidos en los que participó un equipo.
        :param equipo: Nombre del equipo.
        :param condicion: "local", "visitante" o "ambos".
        """
        resultados = []
        with open(self.archivo, mode="r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                if (condicion == "local" and row["Equipo Local"] == equipo) or \
                   (condicion == "visitante" and row["Equipo Visitante"] == equipo) or \
                   (condicion == "ambos" and (row["Equipo Local"] == equipo or row["Equipo Visitante"] == equipo)):
                    resultados.append(row)
        return resultados

    def modificar_partido(self, fecha, equipo_local, equipo_visitante, nuevos_goles_local, nuevos_goles_visitante):
        """
        Modificar el resultado de un partido.
        :param fecha: Fecha del partido.
        :param equipo_local: Equipo local del partido.
        :param equipo_visitante: Equipo visitante del partido.
        """
        partidos = []
        modificado = False

        # Leer y actualizar los datos del archivo
        with open(self.archivo, mode="r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                if row["Fecha-Partido"] == fecha and \
                   row["Equipo Local"] == equipo_local and \
                   row["Equipo Visitante"] == equipo_visitante:
                    row["Goles Local"] = nuevos_goles_local
                    row["Goles Visitante"] = nuevos_goles_visitante
                    modificado = True
                partidos.append(row)

        # Reescribir el archivo con los datos modificados
        with open(self.archivo, mode="w", newline="") as file:
            fieldnames = ["Fecha-Partido", "Equipo Local", "Goles Local", "Equipo Visitante", "Goles Visitante"]
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(partidos)

        return modificado
````

**Ej 5:** Escribir un programa que tome el archivo `torneo.csv` del ejercicio anterior y genere la tabla de posiciones en un archivo en formato json, con nombre `posiciones.json`, con el siguiente formato: `{"posiciones": {"equipo1": {"puntos": ptos_totales, "goles a favor": total_goles_a_favor, "goles_en_contra": total_goles_en_contra}, "equipo2": {"puntos": ptos_totales, "goles a favor": total_goles_a_favor, "goles_en_contra": total_goles_en_contra}, .....}}`. Donde para calcular los puntos por cada partido ganado se otorgan tres puntos, al ganador, y por cada empate un punto a cada equipo. 

````python
import csv
import json

class TablaPosiciones:
    def __init__(self, archivo_torneo="torneo.csv", archivo_posiciones="posiciones.json"):
        self.archivo_torneo = archivo_torneo
        self.archivo_posiciones = archivo_posiciones
        self.posiciones = {}

    def calcular_posiciones(self):
        """Calcula la tabla de posiciones en base a los resultados de los partidos."""
        with open(self.archivo_torneo, mode="r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                equipo_local = row["Equipo Local"]
                equipo_visitante = row["Equipo Visitante"]
                goles_local = int(row["Goles Local"])
                goles_visitante = int(row["Goles Visitante"])

                # Asegurar que los equipos estén en el diccionario de posiciones
                if equipo_local not in self.posiciones:
                    self.posiciones[equipo_local] = {"puntos": 0, "goles a favor": 0, "goles en contra": 0}
                if equipo_visitante not in self.posiciones:
                    self.posiciones[equipo_visitante] = {"puntos": 0, "goles a favor": 0, "goles en contra": 0}

                # Actualizar goles a favor y en contra
                self.posiciones[equipo_local]["goles a favor"] += goles_local
                self.posiciones[equipo_local]["goles en contra"] += goles_visitante
                self.posiciones[equipo_visitante]["goles a favor"] += goles_visitante
                self.posiciones[equipo_visitante]["goles en contra"] += goles_local

                # Asignar puntos en función del resultado
                if goles_local > goles_visitante:
                    self.posiciones[equipo_local]["puntos"] += 3  # Gana equipo local
                elif goles_local < goles_visitante:
                    self.posiciones[equipo_visitante]["puntos"] += 3  # Gana equipo visitante
                else:
                    self.posiciones[equipo_local]["puntos"] += 1  # Empate
                    self.posiciones[equipo_visitante]["puntos"] += 1  # Empate

    def guardar_posiciones_json(self):
        """Guarda la tabla de posiciones en un archivo JSON."""
        tabla_posiciones = {"posiciones": self.posiciones}
        with open(self.archivo_posiciones, mode="w") as file:
            json.dump(tabla_posiciones, file, indent=4)

    def generar_tabla_posiciones(self):
        """Genera la tabla de posiciones y la guarda en formato JSON."""
        self.calcular_posiciones()
        self.guardar_posiciones_json()
        print(f"Tabla de posiciones guardada en {self.archivo_posiciones}")


# Ejemplo de uso
torneo = TablaPosiciones()
torneo.generar_tabla_posiciones()
````

**Ej 6:** Se tienen dos archivos CSV uno con columnas Código de Materia, Legajo, Nota,  y otro con Código de Materia, Departamento. Escribir una función en Python para devuelva un diccionario con clave Departamento y cuyo valor sea la cantidad de alumnos que aprobaron materias de ese Departamento (las materias se aprueban con nota mayor o igual a 4). 

````python
import csv
from collections import defaultdict

def contar_aprobados_por_departamento(archivo_notas, archivo_departamentos):
    # Crear un diccionario para mapear cada código de materia a su departamento
    materias_departamento = {}

    # Leer el archivo de departamentos y crear el mapeo
    with open(archivo_departamentos, mode="r") as file:
        reader = csv.DictReader(file)
        for row in reader:
            codigo_materia = row["Código de Materia"]
            departamento = row["Departamento"]
            materias_departamento[codigo_materia] = departamento

    # Crear un diccionario para contar los aprobados por departamento
    aprobados_por_departamento = defaultdict(int)

    # Leer el archivo de notas y contar aprobados por departamento
    with open(archivo_notas, mode="r") as file:
        reader = csv.DictReader(file)
        for row in reader:
            codigo_materia = row["Código de Materia"]
            nota = int(row["Nota"])

            # Verificar si la nota es aprobatoria y la materia tiene un departamento asignado
            if nota >= 4 and codigo_materia in materias_departamento:
                departamento = materias_departamento[codigo_materia]
                aprobados_por_departamento[departamento] += 1

    return dict(aprobados_por_departamento)
````

````python
# Casos de Prueba
import os

def crear_archivo(nombre, encabezado, datos):
    with open(nombre, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(encabezado)
        writer.writerows(datos)

def test_contar_aprobados_por_departamento_1():
    # Crear archivos de prueba
    archivo_notas = "notas.csv"
    archivo_departamentos = "departamentos.csv"
    crear_archivo(archivo_notas, ["Código de Materia", "Legajo", "Nota"], [
        ["MAT101", "1234", "5"],
        ["MAT101", "5678", "3"],
        ["MAT102", "4321", "6"],
        ["MAT103", "8765", "4"]
    ])
    crear_archivo(archivo_departamentos, ["Código de Materia", "Departamento"], [
        ["MAT101", "Ciencias Exactas"],
        ["MAT102", "Ingeniería"],
        ["MAT103", "Ciencias Exactas"]
    ])

    # Llamar a la función y verificar el resultado
    resultado = contar_aprobados_por_departamento(archivo_notas, archivo_departamentos)
    resultado_esperado = {"Ciencias Exactas": 2, "Ingeniería": 1}

    if resultado == resultado_esperado:
        print("test_contar_aprobados_por_departamento_1 PASSED")
    else:
        print("test_contar_aprobados_por_departamento_1 FAILED")

    # Limpiar archivos de prueba
    os.remove(archivo_notas)
    os.remove(archivo_departamentos)

def test_contar_aprobados_por_departamento_2():
    # Crear archivos de prueba
    archivo_notas = "notas.csv"
    archivo_departamentos = "departamentos.csv"
    crear_archivo(archivo_notas, ["Código de Materia", "Legajo", "Nota"], [
        ["BIO101", "2345", "7"],
        ["BIO102", "6789", "4"],
        ["BIO103", "7890", "2"],
        ["BIO104", "3456", "6"]
    ])
    crear_archivo(archivo_departamentos, ["Código de Materia", "Departamento"], [
        ["BIO101", "Biología"],
        ["BIO102", "Biología"],
        ["BIO104", "Biología"]
    ])

    # Llamar a la función y verificar el resultado
    resultado = contar_aprobados_por_departamento(archivo_notas, archivo_departamentos)
    resultado_esperado = {"Biología": 3}

    if resultado == resultado_esperado:
        print("test_contar_aprobados_por_departamento_2 PASSED")
    else:
        print("test_contar_aprobados_por_departamento_2 FAILED")

    # Limpiar archivos de prueba
    os.remove(archivo_notas)
    os.remove(archivo_departamentos)

# Ejecución de las pruebas
test_contar_aprobados_por_departamento_1()
test_contar_aprobados_por_departamento_2()
````

**Ej 7:**  Crear una base de datos de productos con atributos variables, es decir, donde cada producto pueda tener un conjunto distinto de atributos. Se deben poder listar, buscar un producto, por nombre, o cualquier otro atributo, registrar nuevos productos y eliminarlos. Los datos se deben serializar y persistir en archivo .json

````python
import json
import os

class BaseDeDatosProductos:
    def __init__(self, archivo="productos.json"):
        self.archivo = archivo
        self.productos = {}
        self._cargar_datos()

    def _cargar_datos(self):
        """Carga los datos desde el archivo JSON si existe."""
        if os.path.exists(self.archivo):
            with open(self.archivo, mode="r") as file:
                self.productos = json.load(file)

    def _guardar_datos(self):
        """Guarda los datos en el archivo JSON."""
        with open(self.archivo, mode="w") as file:
            json.dump(self.productos, file, indent=4)

    def registrar_producto(self, nombre, **atributos):
        """Registra un nuevo producto con atributos variables."""
        self.productos[nombre] = atributos
        self._guardar_datos()

    def listar_productos(self):
        """Lista todos los productos."""
        return self.productos

    def buscar_producto(self, **criterios):
        """
        Busca un producto que coincida con los criterios dados.
        :param criterios: Atributos a buscar, como nombre, precio, categoría, etc.
        """
        resultados = []
        for nombre, atributos in self.productos.items():
            if all(atributos.get(k) == v for k, v in criterios.items()):
                resultados.append({nombre: atributos})
        return resultados

    def eliminar_producto(self, nombre):
        """Elimina un producto por su nombre."""
        if nombre in self.productos:
            del self.productos[nombre]
            self._guardar_datos()
            return True
        return False
````

````python
# Ejemplos de uso
db = BaseDeDatosProductos()

# Registrar productos
db.registrar_producto("Laptop", marca="Dell", precio=1200, color="Negro", procesador="Intel i5")
db.registrar_producto("Smartphone", marca="Samsung", precio=800, color="Azul", almacenamiento="128GB")
db.registrar_producto("Tablet", marca="Apple", precio=900, color="Gris", almacenamiento="256GB", peso="400g")

# Listar productos
print("Listado de productos:")
for nombre, atributos in db.listar_productos().items():
    print(nombre, atributos)

# Buscar un producto por nombre
print("\nBuscar producto por nombre 'Laptop':")
print(db.buscar_producto(marca="Dell"))

# Buscar productos por atributos específicos
print("\nBuscar productos de color Azul:")
print(db.buscar_producto(color="Azul"))

# Eliminar un producto
eliminado = db.eliminar_producto("Smartphone")
print("\nProducto 'Smartphone' eliminado:", eliminado)

# Verificar la lista de productos después de la eliminación
print("\nListado de productos después de la eliminación:")
for nombre, atributos in db.listar_productos().items():
    print(nombre, atributos)
````

**Ej 8:** Crear un diccionario que represente a un producto con atributos nombre, precio, descripción y una lista de características. Serializar este diccionario a un archivo JSON. Implementar un menu que permita cargar el diccionario desde memoria, guardar un diccionario y agregar características a un producto existente. Manejar cualquier excepción que pueda surgir.

````python
import json

# Diccionario que representa al producto
producto = {
    "nombre": "Producto X",
    "precio": 0,
    "descripcion": "",
    "caracteristicas": []
}

# Nombre del archivo JSON para persistencia
archivo_json = "producto.json"


def cargar_producto():
    try:
        with open(archivo_json, mode="r") as file:
            global producto
            producto = json.load(file)
        print("Producto cargado exitosamente.")
    except FileNotFoundError:
        print("No se encontró el archivo. Comenzando con un producto vacío.")
    except json.JSONDecodeError:
        print("Error al decodificar el archivo JSON. Verifique el formato.")
    except Exception as e:
        print(f"Error inesperado al cargar el producto: {e}")


def guardar_producto():
    try:
        with open(archivo_json, mode="w") as file:
            json.dump(producto, file, indent=4)
        print("Producto guardado exitosamente.")
    except Exception as e:
        print(f"Error inesperado al guardar el producto: {e}")


def agregar_caracteristica():
    caracteristica = input("Ingrese una característica para el producto: ")
    producto["caracteristicas"].append(caracteristica)
    print("Característica agregada exitosamente.")


def mostrar_menu():
    """Muestra el menú de opciones al usuario."""
    print("\n--- Menú ---")
    print("1. Cargar producto desde archivo")
    print("2. Guardar producto en archivo")
    print("3. Agregar característica al producto")
    print("4. Mostrar producto actual")
    print("5. Salir")


def mostrar_producto():
    """Muestra el producto actual en memoria."""
    print("\nProducto actual:")
    print(json.dumps(producto, indent=4))
````

````python
def main():
    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ")
        
        if opcion == "1":
            cargar_producto()
        elif opcion == "2":
            guardar_producto()
        elif opcion == "3":
            agregar_caracteristica()
        elif opcion == "4":
            mostrar_producto()
        elif opcion == "5":
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida. Intente nuevamente.")

if __name__ == "__main__":
    main()
````

**Ej 9:** Dada la jerarquias de punto y figuras en el plano visto en la segunda clase, escribir una capa de persistencia, sobre json, para que los datos de las figuras creadas se puedan guardar en disco y recuperar del mismo. Todas las figuras se deben guardar en un único archivo. Agregar al menú de main la opción de guardar todas las figuras al disco y de recuperarlas.

````python
import math


class Punto:
    def __init__(self, coord_x=0, coord_y=0):
        """Constructor de la clase punto, por defecto construye
        el punto en el origen"""
        self.x = coord_x
        self.y = coord_y

    def mover(self, incremento_x, incremento_y):
        self.x += incremento_x
        self.y += incremento_y

    def __str__(self):
        """Para imprimir un punto con print"""
        return "(%d, %d)" % (self.x, self.y)


class Figura:
    """En python no existen las clases abstractas, todas las clases son
    concretas"""

    def __init__(self, color_borde=0, color_relleno=0, visibilidad=True):
        self.color_borde = color_borde
        self.color_relleno = color_relleno
        self.visibilidad = visibilidad

    def visible(visibilidad):
        self.visibilidad = visibilidad

    def __str__(self):
        """Con introspeccion puedo preguntar el nombre de la clase y
        el duck typing me permite llamar un método que todavía no está
        definido"""
        return "%s: area: %s" % (self.__class__.__name__, self.area())


class Elipse(Figura):
    def __init__(
        self,
        centro,
        radio_menor,
        radio_mayor,
        color_borde=0,
        color_relleno=0,
        visibilidad=True,
    ):
        super().__init__(color_borde, color_relleno, visibilidad)
        self.centro = centro
        self.radio_menor = radio_menor
        self.radio_mayor = radio_mayor

    def area(self):
        return math.pi * self.radio_menor * self.radio_mayor

    def mover(self, incremento_x, incremento_y):
        self.centro.mover(incremento_x, incremento_y)

    def __str__(self):
        return super().__str__() + "\n\tcentro: " + self.centro.__str__()


class Circulo(Elipse):
    def __init__(self, centro, radio, color_borde=0, color_relleno=0, visibilidad=True):
        super().__init__(centro, radio, radio, color_borde, color_relleno, visibilidad)


class Rectangulo(Figura):
    def __init__(
        self,
        vertice_inf_izq,
        vertice_sup_der,
        color_borde=0,
        color_relleno=0,
        visibilidad=True,
    ):
        super().__init__(color_borde, color_relleno, visibilidad)
        self.vertice_inf_izq = vertice_inf_izq
        self.vertice_sup_der = vertice_sup_der

    def area(self):
        base = self.vertice_sup_der.x - self.vertice_inf_izq.x
        altura = self.vertice_sup_der.y - self.vertice_inf_izq.y
        return base * altura

    def mover(self, incremento_x, incremento_y):
        self.vertice_inf_izq.mover(incremento_x, incremento_y)
        self.vertice_sup_der.mover(incremento_x, incremento_y)

    def __str__(self):
        return (
            super().__str__()
            + "\n\tvertice_inf_izq: "
            + self.vertice_inf_izq.__str__()
            + "\n\tvertice_sup_der: "
            + self.vertice_sup_der.__str__()
        )


class Cuadrado(Rectangulo):
    def __init__(self, vertice_inf_izq, lado, color_borde=0, color_relleno=0, visibilidad=True):
        vertice_sup_der = Punto(vertice_inf_izq.x + lado, vertice_inf_izq.y + lado)
        super().__init__(vertice_inf_izq, vertice_sup_der, color_borde, color_relleno, visibilidad)


class Triangulo(Figura):
    def __init__(self, punto1, punto2, punto3, color_borde=0, color_relleno=0, visibilidad=True):
        super().__init__(color_borde, color_relleno, visibilidad)
        self.p1 = punto1
        self.p2 = punto2
        self.p3 = punto3

    def area(self):
        """[x1*(y3-y2) + x2*(y1-y3) + x3*(y2-y1)]/2"""

        return (
            self.p1.x * (self.p3.y - self.p2.y)
            + self.p2.x * (self.p1.y - self.p3.y)
            + self.p3.x * (self.p2.y - self.p1.y)
        ) / 2

    def mover(self, incremento_x, incremento_y):
        self.p1.mover(incremento_x, incremento_y)
        self.p2.mover(incremento_x, incremento_y)
        self.p3.mover(incremento_x, incremento_y)

    def __str__(self):
        return (
            super().__str__()
            + "\n\tPunto 1: "
            + self.p1.__str__()
            + "\n\tPunto 2: "
            + self.p2.__str__()
            + "\n\tPunto 3: "
            + self.p3.__str__()
        )


def crear_elipse():
    x, y = input("Ingrese centro de la elipse").split()
    x = int(x)
    y = int(y)
    r = float(input("Ingrese radio menor"))
    s = float(input("Ingrese radio mayor"))
    return Elipse(Punto(x, y), r, s)


def crear_circulo():
    x, y = input("Ingrese centro del circulo").split()
    x = int(x)
    y = int(y)
    r = float(input("Ingrese el radio"))
    return Circulo(Punto(x, y), r)


def crear_rectangulo():
    x1, y1 = input("Ingrese vertice inferior izquierdo").split()
    x1 = int(x1)
    y1 = int(y1)
    x2, y2 = input("Ingrese vertice superior derecho").split()
    x2 = int(x2)
    y2 = int(y2)
    return Rectangulo(Punto(x1, y1), Punto(x2, y2))


def crear_cuadrado():
    x1, y1 = input("Ingrese vertice inferior izquierdo").split()
    x1 = int(x1)
    y1 = int(y1)
    lado = float(input("Ingrese lado"))
    return Cuadrado(Punto(x1, y1), lado)


def crear_triangulo():
    x1, y1 = input("Ingrese punto 1 del triangulo").split()
    x1 = int(x1)
    y1 = int(y1)
    x2, y2 = input("Ingrese punto 2 del triangulo").split()
    x2 = int(x2)
    y2 = int(y2)
    x3, y3 = input("Ingrese punto 3 del triangulo").split()
    x3 = int(x3)
    y3 = int(y3)
    return Triangulo(Punto(x1, y1), Punto(x2, y2), Punto(x3, y3))


if __name__ == "__main__":
    switcher = {
        "e": crear_elipse,
        "c": crear_circulo,
        "r": crear_rectangulo,
        "q": crear_cuadrado,
        "t": crear_triangulo,
    }
    figuras = []
    while True:
        print("Crear Figura:")
        print("\te: Elipse")
        print("\tc: Círculo")
        print("\tr: Rectángulo")
        print("\tq: Cuadrado")
        print("\tt: Triángulo")
        op = input("Elija una opción")
        fig = switcher[op]()
        figuras.append(fig)
        continuar = input("Agregar otra figura [s/n]")
        if continuar[0].lower() != "s":
            break

    for fig in figuras:
        print(fig)
````
